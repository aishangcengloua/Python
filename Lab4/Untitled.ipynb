```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib

data = np.loadtxt('Lab4.dat')

def calSSE(X, cidx, K) :
    SSE = 0
    for i in range(K) :
        Ci = np.mean(X[np.where(cidx == i + 1), 0])
        SSE += np.sum(np.square(X[np.where(cidx == i + 1), 0] - Ci))

    return SSE / X.shape[0]

def kmeans(X, K) :
    center_point = []
    for i in range(K) :
        point_x = np.random.uniform(np.min(X, axis = 0)[0], np.max(X, axis = 0)[0])
        point_y = np.random.uniform(np.min(X, axis = 0)[1], np.max(X, axis = 0)[1])
        center_point.append([point_x, point_y])
    center_point = np.array(center_point)
    cluter = np.zeros(X.shape[0]).astype(np.int32)
    item = 5
    while item > 0:
        for i in range(X.shape[0]) :
            distance = center_point
            distance = np.sum(np.square(distance - X[i]), axis = 1)
            cluter[i] = np.argmin(distance) + 1
    
        New_center_point = np.zeros((K, 2))
        
        for i in range(K) :
            New_center_point[i][0] = np.mean(X[np.where(cluter == i + 1), 0])
            New_center_point[i][1] = np.mean(X[np.where(cluter == i + 1), 1])
        if (New_center_point - center_point < 1e-7).all() :
            break
        center_point = New_center_point
        item -= 1

    return cluter, center_point

SSE = []
for K in range(2, 7) :
    cidx, ctrs = kmeans(data, K)
    print(f'K为{K}时的簇点 : {ctrs}')
    SSE.append(calSSE(data, cidx, K))

mark = [ 'r', 'c', 'y', 'k', 'm', 'g', 'g']
plt.ion()
for i in range(K) :
    plt.scatter(data[np.where(cidx == i + 1), 0], data[np.where(cidx == i + 1), 1], marker = '.', color = mark[i])
plt.title('K is 6')
plt.figure()
plt.plot(list(range(2, 7)), SSE, '+--')
plt.ioff()
plt.show()
```
